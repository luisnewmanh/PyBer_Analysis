```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() with a tuple
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() objeact oriented method
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Format
ax.set_xlabel('Date')
ax.set_ylabel('Fare [$]')
ax.set_ylim(0, 45)
ax.set_title("PyBare Fare by Month")
ax.legend()
ax.grid()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Set the x limit between Jan and Jun.
plt.xlim(0, 5)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
# Change size
plt.figure(figsize=(20, 10))
```

```python
#BAR CHARTS
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Bar chart wit format
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# HORIZONTAL BAR CHART
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create plot with transposed axis
plt.barh(y_axis, x_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis)
#invert axis (transpose) and order in ascending months
plt.gca().invert_yaxis()
```

```python
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.xlim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Bar Charts Using the Object-Oriented Approach
# Create the plot with ax.plt()
fig, ax = plt.subplots()
#ax.bar(x_axis, y_axis)
ax.bar(x_axis, y_axis, color="black", label='New York')
ax.legend()
```

```python
# Create the plot with ax.plt() horizontal
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() transposed
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# Create the plot.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
# Format
ax.set_ylabel('Date')
ax.set_xlabel('Fare [$]')
ax.set_xlim(0, 45)
ax.set_ylim(12, -1) #invert axis by switching order
ax.set_title("PyBare Fare by Month")
ax.legend()
```

```python
#SCATTER PLOTS
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color="red", label='Chicago')
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Set the y limit between 0 and 45.
plt.xlim(0, 50)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Add a third dimension by modifying size
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color='xkcd:sky blue', alpha=.8, ec='black', lw=2, s=[i * 5 for i in y_axis], label='Boston')
# Format
ax.set_ylabel('Date')
ax.set_xlabel('Fare [$]')
ax.set_xlim(0, 50)
ax.set_ylim(12, -1) #invert axis by switching order
ax.set_title("PyBare Fare by Month")
ax.legend()
```

```python
# PIE CHARTS
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')
```

```python
#Pie chart oriented method
fig, ax = plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, .2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,labels=x_axis, colors=colors, autopct='%.1f%%', explode=explode_values, shadow='True', startangle=93, counterclock=False)
plt.show()
```
